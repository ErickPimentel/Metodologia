```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sn
import sys
from matplotlib.pyplot import figure


%matplotlib inline
```

```python
df = pd.DataFrame({'faltosos': [8,12,16,7,5,5,14,22,22],
                   'total indivíduos': [46,46,46,26,26,26,37,37,37],
                    'chamada': [2.1,16.45,0.28,0.36,10.51,0.21,1.56,10.26,0.23],
                      'total / cadastro': [3.18,16.45,8.18,1.30,10.51,4.10,2.29,10.26,7.28]},
                   index=['Chamada Oral', 'Lista de Presença', 'Tecnologia Proposta','Chamada Oral', 'Lista de Presença', 'Tecnologia Proposta','Chamada Oral', 'Lista de Presença', 'Tecnologia Proposta'])
df
```

```python
colors = ["red", "red", "red", "blue", "blue", "blue", "green", "green", "green"]
df.plot.bar(y='chamada',figsize=[10,15],color=colors,fontsize=16)
plt.ylabel("Tempo (minutos) ",fontsize=16)                          
plt.xlabel('A                                       B                                       C',fontsize=14)
plt.legend("")
plt.yticks(np.arange(0, 17, 0.5),fontsize=16)
plt.grid()
```

```python
bar_1 = df1["Lista de Presença"]
bar_2 = df1["Chamada Oral"]
bar_3 = df1["Tecnologia Proposta"]
x_pos = np.arange(len(bar_1))
plt.subplots(figsize=(20,10))
first_bar = plt.bar(x_pos, bar_1, 0.5, color='r')
second_bar = plt.bar(x_pos, bar_2, 0.5, color='y')
terc_bar = plt.bar(x_pos, bar_3, 0.5, color='b')
plt.xticks(x_pos, ('Experimento 1','Experimento 2', 'Experimento 3'))
plt.yticks(np.arange(0, 17, 0.5),fontsize=16)
plt.ylabel('Tempo')
plt.legend(["Lista de Presença","Chamada Oral","Tecnologia Proposta"])
plt.grid(True)
```
